```python
import numpy as np
import pandas as pd
import nltk
nltk.download('vader_lexicon')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
test_data= pd.read_csv('/content/drive/MyDrive/Machine-Learning/KaggleKnights/test_absa.csv')
```

```python
import string
def text_cleaner(text):
    text = text.lower()
    text = text.translate(str.maketrans('','',string.punctuation))
    return text
cleaned = lambda x: text_cleaner(x)

test_data["id"]=pd.DataFrame(test_data.text.apply(cleaned))
test_data["predicted"]=pd.DataFrame(test_data.aspect.apply(cleaned))
test_data.head()
```

```python
def sentiment_analyze(text):
    score = SentimentIntensityAnalyzer().polarity_scores(text)
    # print(score)
    predictions = []

    # positive sentiment
    if score['compound'] > 0.1 :
        predictions.append(1)

    # negative sentiment
    elif score['compound'] <= - 0.05 :
        predictions.append(-1)

    # neutral sentiment
    else :
        predictions.append(0)

    predictions = list(map(int,predictions))
    return(predictions)

sen_analyze = lambda x: sentiment_analyze(x)
test_data["predicted"]=pd.DataFrame(test_data.text.apply(sen_analyze))
test_data.head()
```

```python
df = test_data.copy()
df
```

```python
df['id'] = df.index
df['predicted'] = df['predicted'].str.get(0)

new_cols = ["id","text","aspect","predicted"]
df=df[new_cols]
df
```
